```python
import pdfplumber
import pandas as pd

# Define the path to the PDF
pdf_path = r"C:\Users\shrut\Downloads\17134183971722149624file.pdf"

# Open the PDF
with pdfplumber.open(pdf_path) as pdf:
    # Extract tables from specific pages (e.g., pages 112–152)
    unconfined_data = []
    for page_number in range(115, 156):  # 0-indexed
        page = pdf.pages[page_number]
        table = page.extract_table()
        if table:
            unconfined_data.extend(table)

# Convert to a DataFrame
columns = unconfined_data[0]  # Use the first row as column headers
data = unconfined_data[1:]    # Remaining rows are the data
df = pd.DataFrame(data, columns=columns)

# Save to CSV
df.to_csv("unconfined_aquifer_data_new.csv", index=False)
```
